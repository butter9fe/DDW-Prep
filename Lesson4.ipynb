# Python Strings and Tuples

Learn about two important data types: strings for text manipulation and tuples for immutable sequences.

## 1. Strings

A string is a sequence of characters enclosed in quotes (single or double). Strings are:
- **Immutable**: Cannot be changed after creation (you must create a new string)
- **Iterable**: Can loop through each character
- **Indexable**: Can access characters by position
- **Sliceable**: Can extract substrings

Strings are one of the most commonly used data types in Python.

### Basic String Operations

```python
Oshi = "Suisei"
print(Oshi)  # This will print the String "Suisei" as text
```

### Strings are Immutable

You cannot change individual characters in a string. This will raise an error:

```python
# This will raise a TypeError
try:
    Oshi[0] = "H"
except TypeError as e:
    print(f"Error: {e}")
    print("Strings are immutable - you cannot change individual characters!")
```

### Iterating Through Strings

```python
Oshi = "Suisei"

# Strings are Iterable - loop through each character
for character in Oshi:
    print(character)
# This will print each character in the string "Suisei" on a new line
```

### String Indexing and Slicing

You can access parts of a string using indices and slices.

```python
Oshi = "Suisei"

# Indexing - access a single character
print(f"First character: {Oshi[0]}")  # S
print(f"Second character: {Oshi[1]}")  # u
print(f"Last character: {Oshi[-1]}")  # i (negative index counts from the end)

# Slicing - extract a substring
print(f"Characters from index 1 to 4: {Oshi[1:4]}")  # uis (end index is exclusive)
print(f"First 3 characters: {Oshi[:3]}")  # Sui
print(f"Characters from index 2 onwards: {Oshi[2:]}")  # isei
```

### String Concatenation

Combine strings using the `+` operator.

```python
Oshi = "Suisei"

# String concatenation
Oshi = Oshi + " is my favorite VTuber."
print(Oshi)  # Suisei is my favorite VTuber.
```

### String Methods

Strings have many built-in methods for manipulation. Here are some commonly used ones:

```python
Oshi = "Suisei is my favorite VTuber."

# Convert to uppercase
print(Oshi.upper())  # SUISEI IS MY FAVORITE VTUBER.

# Convert to lowercase
print(Oshi.lower())  # suisei is my favorite vtuber.

# Replace text
print(Oshi.replace("Suisei", "Miko"))  # Miko is my favorite VTuber.

# Find the index of a substring
print(Oshi.find("favorite"))  # 16

# Check if string starts with a substring
print(Oshi.startswith("Suisei"))  # True

# Check if string ends with a substring
print(Oshi.endswith("VTuber."))  # True

# Split string into a list
words = Oshi.split(" ")
print(words)  # ['Suisei', 'is', 'my', 'favorite', 'VTuber.']
```

### String Formatting

There are several ways to format strings with variables.

```python
name = "Suisei"
age = 18
height = 160.5

# Method 1: F-strings (Modern and recommended)
print(f"{name} is {age} years old and {height} cm tall.")

# Method 2: .format() method
print("{} is {} years old and {} cm tall.".format(name, age, height))

# Method 3: String concatenation
print(name + " is " + str(age) + " years old and " + str(height) + " cm tall.")
```

## 2. Tuples

A tuple is an ordered collection of items enclosed in parentheses `()`. Tuples are:
- **Immutable**: Cannot be changed after creation (like strings)
- **Ordered**: Items have a defined order and can be accessed by index
- **Can contain mixed types**: Like lists, tuples can contain any data type
- **Iterable**: Can loop through items

Tuples are lighter weight than lists and are used when data shouldn't change.

### Basic Tuple Operations

```python
numbers = (88, 1, 20)
print(numbers)  # This will print the tuple (88, 1, 20)
```

### Tuples are Immutable

You cannot change, add, or remove elements from a tuple after creation.

```python
numbers = (88, 1, 20)

# This will raise a TypeError because tuples do not support item assignment
try:
    numbers[0] = 99
except TypeError as e:
    print(f"Error: {e}")
    print("Tuples are immutable - you cannot change elements!")
```

### Tuples with Mixed Data Types

```python
# Similar to lists, tuples can contain ANY data type
StudioStellar = ("Suisei", 88, True, 160.5)
print(StudioStellar)  # This will print the tuple ("Suisei", 88, True, 160.5)

# Access elements by type
print(f"Name (string): {StudioStellar[0]}")
print(f"Number (int): {StudioStellar[1]}")
print(f"Boolean: {StudioStellar[2]}")
print(f"Height (float): {StudioStellar[3]}")
```

### Iterating Through Tuples

```python
StudioStellar = ("Suisei", 88, True)

# Tuples are iterable (Similar to Lists)
for item in StudioStellar:
    print(item)
# This will print each item in the tuple, each on a new line
```

### Tuple Indexing and Slicing

```python
StudioStellar = ("Suisei", 88, True, 160.5)

# Tuples also support indexing
print(f"First item: {StudioStellar[0]}")  # Suisei
print(f"Last item: {StudioStellar[-1]}")  # 160.5

# Tuples also support slicing
print(f"Items from index 1 to 3: {StudioStellar[1:3]}")  # (88, True)
```

### Important: Single Element Tuples

Single element tuples MUST have a trailing comma, otherwise Python treats them as regular values.

```python
# Single element tuple - requires a comma
single_element_tuple = (88,)
print(f"Type: {type(single_element_tuple)}")
print(f"Value: {single_element_tuple}")  # (88,)

# Without the comma - this is NOT a tuple
not_a_tuple = (88)
print(f"\nType: {type(not_a_tuple)}")
print(f"Value: {not_a_tuple}")  # 88 (just an integer)

# Remember: Single element tuples need a comma!
```

### Tuples vs Lists

| Feature | Tuple | List |
|---------|-------|------|
| Mutable | ❌ No (immutable) | ✅ Yes (mutable) |
| Syntax | `()` parentheses | `[]` square brackets |
| Performance | Faster | Slower |
| Use Case | Fixed data, function returns | Dynamic data, frequent changes |
| Methods | Limited | Many modification methods |

**Use tuples when:** Data shouldn't change, performance matters, or you need to use items as dictionary keys.

**Use lists when:** You need to add, remove, or modify elements.

### Tuple Unpacking

A powerful feature of tuples is unpacking - assigning tuple elements to individual variables.

```python
# Tuple unpacking
person = ("Suisei", 18, 160.5)
name, age, height = person

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Height: {height}")

# This is the same as:
# name = person[0]
# age = person[1]
# height = person[2]
```
