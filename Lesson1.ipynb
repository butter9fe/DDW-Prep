## Datatypes

Python supports several fundamental data types:
- **String**: Text data enclosed in quotes

You can check the data type of any variable using the `type()` function.

### String Datatype Example

```python
player_name = "Steve"
print(player_name, type(player_name))
```

```python
greeting = "Steve says 'Minecraft is fun!' and chops a tree."
print(greeting)
```

```python
story = """Once upon a time,
there was a Minecraft player named Steve.
He chopped some trees and built a house."""
print(story)
```

```python
has_weapon = True
print(has_weapon, type(has_weapon))

has_armour = False
print(has_armour, type(has_armour))

print(has_weapon == 1) # True!
```

### Integer Datatype Example

```python
diamonds = 18
print(diamonds, type(diamonds))
```

### Float Datatype Example

```python
walk_speed = 4.5
print(walk_speed, type(walk_speed))
```

## Arithmetic

Arithmetic operations are used to perform mathematical calculations. Python supports:
- **Addition**: `+`
- **Subtraction**: `-`
- **Multiplication**: `*`
- **Division**: `/`
- **Exponentiation**: `**`
- **Modulo**: `%` (remainder after division)

Operations follow the standard order of operations (PEMDAS).

### Arithmetic Example 1: Mixed Operations

```python
vents = 1 + 2 * 3 - 4 / 5
print(vents)
```

### Arithmetic Example 2: Exponentiation and Modulo

```python
sabotages = 2 ** 3 % 8
print(sabotages)
```

## Comparisons

Comparison operators are used to compare two values. They return Boolean values (True or False):
- **Equal to**: `==`
- **Not equal to**: `!=`
- **Greater than**: `>`
- **Less than**: `<`
- **Greater than or equal to**: `>=`
- **Less than or equal to**: `<=`

### Comparison Examples

```python
imposters = 5
crewmates = 10

print(f"imposters == crewmates: {imposters == crewmates}") # Is imposters = crewmates?
print(f"imposters != crewmates: {imposters != crewmates}") # Is imposters not equal crewmates?
print(f"imposters < crewmates: {imposters < crewmates}") # Is imposters less than crewmates?
print(f"imposters >= crewmates: {imposters >= crewmates}") # Is imposters greater than or equal to crewmates?
print(f"imposters <= crewmates: {imposters <= crewmates}") # Is imposters less than or equal to crewmates?
```

## Logical Operators

Logical operators are used to combine conditional statements:
- **and**: Returns True if both statements are true
- **or**: Returns True if at least one statement is true
- **not**: Returns the opposite boolean value

### Logical Operator Examples

```python
imposters = 2
crewmates = 15

# AND operator
a = imposters == 10 and crewmates == 10
print(f"imposters == 10 and crewmates == 10: {a}") # Are both imposters and crewmates = 10?

# OR operator
b = imposters == 2 or crewmates == 20 # Is either imposters = 2 or crewmates = 20?
print(f"imposters == 2 or crewmates == 20: {b}")

# NOT operator
c = not imposters > crewmates # Is it not true that imposters is greater than crewmates?
print(f"not imposters > crewmates: {c}")
```

### Chained Comparator Examples

```python
d = 10 <= crewmates <= 20
print(f"Is the number of crewmates between 10 and 20? {d}")

# You can also do it in the reverse direction
e = 20 >= crewmates >= 10
print(f"Is the number of crewmates between 10 and 20? {e}")

# AVOID THIS! It can make your logic hard to debug
f = 10 <= crewmates >= 20
print(f"Is the number of crewmates more than 10 and 20? {f}")
```

## Lists

A list is a collection of items in a specific order. Lists:
- Are created using square brackets `[]`
- Can contain different data types
- Are indexed starting from 0
- Support membership testing with `in` and `not in`

### List Examples

```python
player_info = ["Steve", True, 18, 4.5]
print(player_info)

# Access the first element (index 0)
print(player_info[0])
```

### Membership Testing with Lists

```python
player = "Steve"
print(f"'{player}' in player_info: {player in player_info}")

mob = "Creeper"
print(f"'{mob}' not in player_info: {mob not in player_info}")
```

## Variable Assignments

Variable assignment is how we store values in memory:
- **Basic assignment**: Assign a value to a variable
- **Update assignment**: Modify a variable's value
- **Simultaneous assignment**: Assign multiple values at once
- **Swapping**: Exchange values between variables

### Basic and Update Assignment

```python
imposters = 1
print(f"imposters initial value: {imposters}")

imposters = imposters + 1
print(f"imposters after adding 1: {imposters}")
```

### Simultaneous Assignment and Swapping

```python
imposters, crewmates =1, 2
print(f"Before swap: imposters = {imposters}, crewmates = {crewmates}")

imposters, crewmates = crewmates, imposters
print(f"After swap: imposters = {imposters}, crewmates = {crewmates}")
```

## Conditionals (if/elif/else)

Conditional statements allow your program to make decisions:
- **if**: Execute code if a condition is True
- **elif**: Execute code if the previous condition was False and this condition is True
- **else**: Execute code if all previous conditions were False

You can also nest conditionals for more complex logic.

### Conditional Example 1: Age Classification

```python
age = 17

if age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
elif age < 65:
    print("You are an adult.")
else:
    print("You are a senior.")
```

### Conditional Example 2: Nested Conditions

```python
age = 15
in_school = True

if 13 <= age <= 19:
    if in_school:
        print("You are a teenage student.")
    else:
        print("You are a teenager but not in school")
else:
    print("You are not a teenager.")
```

### Conditional Example 3: Multiple Conditions

```python
is_raining = True
is_cold = True

# Using multiple if statements
print("Option 1: Multiple if statements")
if is_raining:
    print("Take an umbrella.")
if is_cold:
    print("Wear a warm coat.")

# Using if/elif
print("\nOption 2: Using if/elif")
if is_raining:
    print("Take an umbrella.")
elif is_cold:
    print("Wear a warm coat.")
```

## Loops

Loops allow you to execute code repeatedly:
- **for loop**: Iterate over a sequence (list, range, etc.)
- **break**: Exit the loop prematurely
- **continue**: Skip to the next iteration

The `for` loop is commonly used with the `range()` function to iterate a specific number of times.

### Loop Example with break Statement

```python
Energy = 100

# Loop through each day of the week
for day in range(1, 8):
    print(f"Day {day}: Energy level is {Energy}.")
    Energy -= 20  # Decrease energy by 20 each day

    if Energy <= 0:
        break

day += 1  # Increment day to reflect the day when energy is depleted
print(f"Day {day}: Energy is depleted. Time to rest.")
```

## The Range Function

The `range()` function generates a sequence of numbers:
- **range(stop)**: Generates numbers from 0 to stop-1
- **range(start, stop)**: Generates numbers from start to stop-1
- **range(start, stop, step)**: Generates numbers from start to stop-1, incrementing by step

The stop value is always exclusive (not included in the sequence).

For more info: https://docs.python.org/3/library/stdtypes.html#range

### Range Example with Step

```python
print("Using range(1, 8, 2):")
for day in range(1, 8, 2):
    # Start at 1,
    # end at 8 (non-inclusive),
    # incremented by 2
    print(day)
```

### Additional Range Examples

```python
# Example 1: range(stop)
print("range(5):")
for i in range(5):
    print(i)

# Example 2: range(start, stop)
print("\nrange(2, 7):")
for i in range(2, 7):
    print(i)
```
