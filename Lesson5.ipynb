# Python Dictionaries

Learn how to work with dictionaries - Python's most powerful data structure for storing and organizing data using key-value pairs.

## Introduction to Dictionaries

A dictionary is a collection of key-value pairs enclosed in curly braces `{}`. Dictionaries:
- Store data as pairs: `key: value`
- Are **mutable** (can be modified)
- Use **keys** to access values (not indices like lists)
- Have **unordered** items (in Python 3.7+, insertion order is preserved)
- Are **indexed** by keys (which must be unique)

Dictionaries are ideal for storing related information together with descriptive labels.

### Basic Dictionary Operations

```python
# Creating a simple dictionary
d = {"apple": 3, "orange": 2, "banana": 1}
print(d)
```

### Accessing Dictionary Values

There are two main ways to access values in a dictionary.

```python
d = {"apple": 3, "orange": 2, "banana": 1}

# Method 1: Using square brackets []
print(f"Apples: {d['apple']}")  # Output: 3

# Method 2: Using .get() method
print(f"Oranges: {d.get('orange')}")  # Output: 2

# .get() with a default value (safer!)
print(f"Mangoes: {d.get('mango', 'N/A')}")  # Output: N/A

# Why is .get() better?
# If you use d['mango'] and 'mango' doesn't exist, you'll get a KeyError
# With .get(), you can provide a default value instead
print(f"\n.get() is safer when keys might not exist!")
```

### Nested Dictionaries

Dictionaries can contain other dictionaries, allowing you to store complex, hierarchical data.

```python
# Dictionary containing dictionaries
students = {
    "alice": {"age": 20, "grade": "A"},
    "bob": {"age": 21, "grade": "B"}
}

# Accessing nested values
print(f"Alice's age: {students['alice']['age']}")  # Output: 20
print(f"Bob's grade: {students['bob']['grade']}")  # Output: B

# Print all students
print(f"\nAll students: {students}")
```

## Modifying Dictionaries

Dictionaries are mutable, so you can add, update, and remove key-value pairs.

### Adding and Updating Keys

```python
student = {"name": "Alice", "age": 20}
print(f"Initial: {student}")

# Update existing key
student["age"] = 22
print(f"After updating age: {student}")

# Add a new key
student["gender"] = "Female"
print(f"After adding gender: {student}")

# Add multiple keys
student["grade"] = "A"
print(f"After adding grade: {student}")
```

### Deleting Keys

```python
student = {"name": "Alice", "age": 22, "grade": "A", "gender": "Female"}
print(f"Initial: {student}")

# Method 1: Using del
del student["grade"]
print(f"After del student['grade']: {student}")

# Method 2: Using .pop() - removes and returns the value
removed_gender = student.pop("gender")
print(f"\nRemoved gender: {removed_gender}")
print(f"After pop: {student}")

# Method 3: Using .popitem() - removes the last inserted item
student["city"] = "New York"
print(f"\nBefore popitem: {student}")
last_item = student.popitem()
print(f"Removed last item: {last_item}")
print(f"After popitem: {student}")
```

## More Dictionary Methods

Dictionaries have several useful built-in methods for accessing different parts of the dictionary.

### .keys(), .values(), and .items()

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# .keys() - returns all keys
print(f"Keys: {student.keys()}")
# Output: dict_keys(['name', 'age', 'grade'])

# .values() - returns all values
print(f"\nValues: {student.values()}")
# Output: dict_values(['Alice', 20, 'A'])

# .items() - returns all key-value pairs as tuples
print(f"\nItems: {student.items()}")
# Output: dict_items([('name', 'Alice'), ('age', 20), ('grade', 'A')])
```

### Other Useful Dictionary Methods

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# .clear() - removes all items
# student.clear()  # Uncomment to see it in action

# .copy() - creates a shallow copy
student_copy = student.copy()
print(f"Copy: {student_copy}")

# .update() - adds/updates multiple key-value pairs
student.update({"age": 21, "city": "New York"})
print(f"\nAfter update: {student}")

# .setdefault() - gets a value or sets a default if key doesn't exist
phone = student.setdefault("phone", "N/A")
print(f"\nPhone (using setdefault): {phone}")
print(f"Dictionary after setdefault: {student}")

# Check if key exists
print(f"\n'age' in student: {'age' in student}")
print(f"'email' in student: {'email' in student}")
```

## Iterating Through Dictionaries

There are multiple ways to loop through a dictionary, each useful for different situations.

### Iterating Through Keys

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# Method 1: Direct iteration (iterates through keys)
for key in student:
    print(f"{key}: {student[key]}")

# Method 2: Using .keys() explicitly
print("\nUsing .keys():")
for key in student.keys():
    print(f"{key}: {student[key]}")

# Expected output:
# name: Alice
# age: 20
# grade: A
```

### Iterating Through Values

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# Using .values() - only get the values
for value in student.values():
    print(value)

# Expected output:
# Alice
# 20
# A
```

### Iterating Through Key-Value Pairs (Recommended)

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# Using .items() with tuple unpacking - most Pythonic way
for key, value in student.items():
    print(f"{key}: {value}")

# Expected output:
# name: Alice
# age: 20
# grade: A

print("\n.items() is the most readable and recommended approach!")
```

### Practical Example: Character Frequency Counter

A common use case for dictionaries is counting occurrences.

```python
# Count character frequencies in a string
text = "banana"
freq = {}

for char in text:
    # .get(char, 0) returns the count if it exists, or 0 if it doesn't
    freq[char] = freq.get(char, 0) + 1

print(f"Original text: {text}")
print(f"Character frequencies: {freq}")
# Expected output: {'b': 1, 'a': 3, 'n': 2}

# Display results nicely
print("\nCharacter frequencies:")
for char, count in freq.items():
    print(f"'{char}' appears {count} time(s)")
```

## Practical Examples

Let's explore real-world use cases for dictionaries.

### Example 1: Student Grades Management

```python
# Managing student grades
students = {
    "Alice": {"math": 95, "english": 88, "science": 92},
    "Bob": {"math": 87, "english": 91, "science": 85},
    "Charlie": {"math": 92, "english": 89, "science": 94}
}

# Find the average grade for each student
for student_name, grades in students.items():
    average = sum(grades.values()) / len(grades)
    print(f"{student_name}'s average grade: {average:.2f}")

# Find the student with the highest math grade
highest_math = max(students.items(), key=lambda x: x[1]["math"])
print(f"\nHighest math grade: {highest_math[0]} with {highest_math[1]['math']}")
```

### Example 2: Product Inventory

```python
# Product inventory
inventory = {
    "apples": 50,
    "oranges": 30,
    "bananas": 25,
    "grapes": 40
}

# Add new products
inventory["mangoes"] = 15

# Update inventory
inventory["apples"] -= 10  # Sell 10 apples

# Display inventory
print("Current Inventory:")
total = 0
for product, quantity in inventory.items():
    print(f"{product.capitalize()}: {quantity}")
    total += quantity
print(f"Total items: {total}")
```
