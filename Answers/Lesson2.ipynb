# Python Functions: Return Statements

Learn how to use return statements in Python functions to create reusable, efficient code.

## What are Functions?

A function is a reusable block of code that performs a specific task. Functions:
- Are defined using the `def` keyword
- Can accept parameters (inputs)
- Can return values (outputs)
- Help organize and simplify code

The basic syntax is:
```python
def function_name(parameters):
    # function body
    return result
```

## The Return Statement

The `return` statement is used to:
- Send a value back from a function to the caller
- Exit the function immediately
- Store the result in a variable for later use

**Important**: `print()` and `return` are NOT the same!
- `print()` displays output to the console but doesn't return a value
- `return` sends a value back to the caller and can be stored in a variable

## Example 1: Print vs Return

Let's see the difference between using `print()` and `return` in a function.

### Version 1: Using print() instead of return

````python
def multiply(a, b) -> int:
    """Returns the product of a and b."""
    print(a * b)

result = multiply(3, 4)
print("result is: ", result)
````

**What happens?** 
- The function prints `12` to the console
- But `result` is set to `None` because the function doesn't return a value
- You can't use the calculated value later in your code

### Version 2: Using return (Correct Way)

````python
def multiply(a, b) -> int:
    """Returns the product of a and b."""
    return a * b

result = multiply(3, 4)
print("result is: ", result)
````

**What happens?**
- The function calculates `a * b` and returns the value
- `result` is set to `12`
- The returned value can be used in further calculations or operations

## Example 2: Practical Applications of Return

Return values become even more useful when you need to use calculated values in your code.

### Quadratic Function Example

Let's create a function that calculates a quadratic expression: y = x² + 5x + 4

### Version 1: Inefficient Approach

````python
def quadratic2(x) -> int:
    y = x*x + 5*x + 4
    return y

a = quadratic2(3)
b = float(a)  # Lines 5-6 are inefficient and time-consuming
print(b)
````

**Why is this inefficient?**
- We calculate the quadratic function and store it in variable `a`
- Then we convert it to float and store it in variable `b`
- This requires two steps and creates unnecessary intermediate variables

### Version 2: Efficient Approach

````python
def quadratic3(x) -> int:
    y = x*x + 5*x + 4
    return y

# We can do this in one line because the function returns a value
b = float(quadratic3(1))
print(b)
````

**Why is this more efficient?**
- We call the function and convert the result to float in a single line
- No intermediate variables needed
- Code is cleaner and easier to read
- The return value can be immediately used in another function call

## Advantages of Using Return

**Reusability**: Return values can be used multiple times in your code

**Composition**: Return values can be passed directly to other functions

**Flexibility**: You can store, manipulate, or transform the returned value

**Clarity**: Code with return statements is easier to understand and maintain

**Testing**: Functions with return values are easier to test and debug

## Key Points to Remember

- **print() vs return**: print() shows output, return() sends a value back
- **Multiple returns**: A function can have multiple return statements (but only one will execute)
- **Return None**: If a function doesn't explicitly return a value, it returns `None`
- **Return type hints**: The `-> type` annotation indicates what type of value the function returns
- **Function composition**: Return values allow you to chain functions together efficiently

### Why Variable Scope Matters

1. **Avoiding conflicts**: Local variables don't interfere with global variables of the same name
2. **Data isolation**: Function parameters and local variables are isolated within the function
3. **Memory efficiency**: Local variables are created when the function is called and deleted when it exits
4. **Code clarity**: It's clear which variables are used where, making code easier to understand
5. **Debugging**: Scope issues can cause hard-to-find bugs if not understood properly

**What happens?**
- The global variable `message` is defined outside the function with value "Hello from the global scope!"
- Inside the `greet()` function, we create a NEW local variable also named `message` with a different value
- When we call `greet()`, it prints the local version: "Hello from the local scope!"
- Outside the function, the global version is still unchanged: "Hello from the global scope!"
- **Key point**: The local variable shadows the global variable inside the function, but they are separate variables

## Global vs Local Scope

Variables in Python have different scopes depending on where they are defined:
- **Global scope**: Variables defined outside of any function (accessible everywhere)
- **Local scope**: Variables defined inside a function (only accessible within that function)

Understanding variable scope is crucial for writing clear, bug-free code.

### Global vs Local Scope Example

````python
message = "Hello from the global scope!"

def greet():
    message = "Hello from the local scope!"
    print("Inside function:", message)

greet()
print("Outside function:", message)
````

### Question 2: Function vs Print
Write two functions that do the same calculation but one uses `print()` and one uses `return()`:
- `double_with_print(x)`: Prints x * 2 but doesn't return anything
- `double_with_return(x)`: Returns x * 2

Test both functions and explain why storing the result from `double_with_print()` doesn't work.

````python
# Question 2: Function vs Print
# Compare functions that use print() vs return()

def double_with_print(x):
    # This function prints but doesn't return anything
    print(x * 2)

def double_with_return(x):
    # This function returns the value
    return x * 2

# Test both functions
print("Using print():")
result1 = double_with_print(5)
print(f"Stored value: {result1}")

print("\nUsing return():")
result2 = double_with_return(5)
print(f"Stored value: {result2}")
````

### Question 3: Function with Multiple Parameters
Create a function `calculate_area()` that takes length and width as parameters and returns the area of a rectangle. Test it with length=10 and width=5.

````python
# Question 3: Function with Multiple Parameters
# Write a function that calculates the area of a rectangle

def calculate_area(length, width):
    # Multiply length and width to get the area
    area = length * width
    return area

# Test with length=10 and width=5
area_result = calculate_area(10, 5)
print(f"Area of rectangle with length 10 and width 5: {area_result}")
````

### Question 4: Conditional Returns
Write a function `check_age()` that takes an age parameter and returns "Adult" if age >= 18, otherwise returns "Minor".

````python
# Question 4: Conditional Returns
# Write a function that returns different values based on age

def check_age(age):
    # Check if age is 18 or older
    if age >= 18:
        return "Adult"
    else:
        return "Minor"

# Test with different ages
print(f"Age 20: {check_age(20)}")
print(f"Age 15: {check_age(15)}")
````

### Question 5: Function Chaining
Write a function `square(x)` that returns x squared, and another function `double(x)` that returns x * 2. Test chaining these functions: `square(double(3))`.

````python
# Question 5: Function Chaining
# Write two functions and chain them together

def square(x):
    # Return x squared
    return x ** 2

def double(x):
    # Return x multiplied by 2
    return x * 2

# Test chaining: double(3) = 6, then square(6) = 36
result = square(double(3))
print(f"square(double(3)) = {result}")
````
